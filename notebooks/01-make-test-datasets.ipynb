# Make test datasets for processing multisearch results

```python
import pandas as pd
```

```python
def _make_multisearch_csv(
    moltype,
    ksize,
    pipeline_outdir="s3://seanome-kmerseek/scope-benchmark/pipeline-outputs/2024-10-08__protein_k5-20",
    query="astral-scopedom-seqres-gd-sel-gs-bib-40-2.08.part_001.fa",
    against="astral-scopedom-seqres-gd-sel-gs-bib-40-2.08.part_001.fa",
):
    basename = f"{query}--in--{against}.{moltype}.{ksize}.multisearch.csv"
    csv = f"{pipeline_outdir}/sourmash/multisearch/{basename}"
    return csv


csv = _make_multisearch_csv("protein", 10)
csv
```

```python
multisearch = pd.read_csv(csv)
print(multisearch.shape)
multisearch.head()
```

### Remove unused categories for testing

```python
from typing import get_args

from scop_constants import SCOP_LINEAGES
```

```python
lineages = get_args(SCOP_LINEAGES)
```

```python
multisearch_parsed_class_a.query_superfamily.cat.remove_unused_categories()
```

```python
query_match = "query", "match"

for lineage in lineages:
    for qm in query_match:
        col = f"{qm}_{lineage}"
        multisearch_parsed_class_a[col] = multisearch_parsed_class_a[
            col
        ].cat.remove_unused_categories()
        print(multisearch_parsed_class_a[col].cat.categories)
```

```python
pq = "test-data/process_scop_sourmash_multisearch/multisearch_output_multisearch_results_processed_filtered.pq"

multisearch_parsed_class_a.to_parquet(
    pq,
    index=False,
    engine="pyarrow",
)
```
